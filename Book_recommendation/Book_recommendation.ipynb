# Most Popular Books

```python
popular_df=rating_df.merge(book_df,on='ISBN')
popular_df
```

```python
new_df=popular_df.drop_duplicates(subset=['Book-Title','ISBN'])
new_df=new_df[['Book-Title','Book-Author','Year-Of-Publication','Publisher','Image-URL-M']]
new_df
```

```python
total_readings=popular_df.groupby('Book-Title').count().sort_values('User-ID',ascending=False)
total_readings.rename(columns={'User-ID':'Times readed'},inplace=True)
total_readings=total_readings.reset_index()
total_readings=total_readings[['Book-Title','Times readed']].head(200)
total_readings
```

```python
book_rating_df=popular_df.groupby('Book-Title').mean().reset_index()[['Book-Title','Book-Rating']]
book_rating_df
```

```python
total_readings=pd.merge(total_readings,book_rating_df,on='Book-Title')
total_readings=total_readings.sort_values('Book-Rating',ascending=False)
total_readings
```
